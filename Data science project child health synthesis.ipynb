# Change interval by underweight, normal weight, overweight and obesity

```python
health = ['Underweight', 'Normal weight', 'Overweight', 'Obesity']
df1['my_cut'] = pd.cut(x = df1['index'], bins = [1.0, 18.5, 25.0, 30.0, 45.0], labels = health)
```

```python
from sklearn import svm
from sklearn.metrics import confusion_matrix, classification_report, accuracy_score
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
```

```python
x = df1.drop('my_cut', axis = 1)
y = df1['my_cut']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
s = StandardScaler()
x_train = s.fit_transform(x_train)
x_test = s.transform(x_test)
```

```python
clt = svm.SVC()
clt.fit(x_train, y_train)
pred = clt.predict(x_test)
```

```python
print(classification_report(y_test, pred))
```

```python
print(confusion_matrix(y_test, pred))
```

```python
print(accuracy_score(y_test, pred))
```
